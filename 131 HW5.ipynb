```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
regions = []
for i in range(10000):
    region = [np.random.uniform(-10,0), np.random.uniform(0,10)]
    regions.append(region)
```

```python
def find_consequential_probability(x, stimulus, regions):
    stimulus_count = 0
    x_count = 0
    for i in regions:
        if (i[0] <= stimulus <= i[1]):
            stimulus_count += 1/(abs(i[1] - i[0]))
            if (i[0] <= x <= i[1]):
                x_count += 1/(abs(i[1] - i[0]))
    return x_count/stimulus_count
```

# Question 1

```python
one_given_zero = (find_consequential_probability(1, 0, regions))*100
print(f'The probability that x=1 is in the concept given zero is in the concept is {one_given_zero}.')
```

# Question 2

```python
x_data = np.arange(-10,10.01,.01)
y_data = []
for i in range(len(x_data)):
    y_data.append(find_consequential_probability(x_data[i], 0, regions))


plt.plot(x_data, y_data)
```

This graph makes sense because the farther you get from the stimulus(0), the lower the probability is that you'll get the extremes in the consequential region. Further, since we compute height as 1/(length of each consequential region), we get this graph peaking at (0,1) since 0 is in every consequential region.

Plotting a logarithmic y axis can help us deduce if the curve is exponential. Taking the log of an exponential function results in the line y=x. ln(e^x) = x. Thus, if we take the log of the curve and get a line, then the curve must be exponential.

```python
plt.plot(x_data,y_data)
plt.yscale('log')
plt.xlim(-5,5)
plt.ylim(.1,1)
```

At first, the log graph from ranges (-10,10) of our curve presents non-exponential, as there is no indication of a linear output on a logarithmic axis.

However, upon closer inspection of the graph from ranges (-5,5), we get an an increasing line peaking at 0, then a decreasing line. This tells us that our curve is in fact exponential from (-5,5).

This implies that the rate of change of probabilities from (-5,5) is much faster than from (-10,-5) or (5,10).

We notice how as we add more consquential regions, the curve gets much smoother. In other words, the more consequential regions there are, the more they will smoothly cover all the points.
